```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

# Initialize SimpleImputer to replace NaN with the mean of each column
imputer = SimpleImputer(strategy='mean')

# You can create a pipeline to handle missing values and then fit the model
lr = make_pipeline(imputer, LinearRegression())
lr.fit(x_train, y_train)

rf = make_pipeline(imputer, RandomForestRegressor())
rf.fit(x_train, y_train)

gb = make_pipeline(imputer, GradientBoostingRegressor())
gb.fit(x_train, y_train)

xg = make_pipeline(imputer, XGBRegressor())
xg.fit(x_train, y_train)
```

```python
y_pred1 = lr.predict(x_test)
y_pred2 = rf.predict(x_test)
y_pred3 = gb.predict(x_test)
y_pred4 = xg.predict(x_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({
    'Models': ['LR', 'RF', 'GBR', 'XG'],
    'R2_SCORE': [score1, score2, score3, score4]
})
```

```python
from tkinter import *
import joblib
import pandas as pd

# Function to show the entry fields and make predictions
def show_entry_fields():
    # Retrieve input values from the entry fields
    p1 = float(e1.get())
    p2 = float(e2.get())
    p3 = float(e3.get())
    p4 = float(e4.get())
    p5 = float(e5.get())
    p6 = float(e6.get())
    p7 = float(e7.get())
    
    # Load the pre-trained model
    model = joblib.load('car_price_predictor')

    # Create a DataFrame with the input values
    data_new = pd.DataFrame({
        'Present_Price': [p1],
        'Driven_kms': [p2],
        'Fuel_Type': [p3],
        'Selling_type': [p4],
        'Transmission': [p5],
        'Owner': [p6],
        'Age': [p7]
    })
    
    # Predict the car price using the model
    result = model.predict(data_new)

    # Display the result in the Tkinter window
    label_result.config(text=f"Car Purchase Amount: {result[0]:.2f}")
    print(f"Car Purchase Amount: {result[0]:.2f}")

# Setting up the Tkinter GUI
master = Tk()
master.title("Car Price Prediction using Machine Learning")

# Title Label
Label(master, text="Car Price Prediction using Machine Learning", bg="black", fg="white").grid(row=0, columnspan=2)

# Labels for input fields
Label(master, text="Present_Price").grid(row=1)
Label(master, text="Driven_kms").grid(row=2)
Label(master, text="Fuel_Type").grid(row=3)
Label(master, text="Selling_type").grid(row=4)
Label(master, text="Transmission").grid(row=5)
Label(master, text="Owner").grid(row=6)
Label(master, text="Age").grid(row=7)

# Entry fields for user input
e1 = Entry(master)
e2 = Entry(master)
e3 = Entry(master)
e4 = Entry(master)
e5 = Entry(master)
e6 = Entry(master)
e7 = Entry(master)

# Grid placement for the entry fields
e1.grid(row=1, column=1)
e2.grid(row=2, column=1)
e3.grid(row=3, column=1)
e4.grid(row=4, column=1)
e5.grid(row=5, column=1)
e6.grid(row=6, column=1)
e7.grid(row=7, column=1)

# Button to trigger prediction
Button(master, text='Predict', command=show_entry_fields).grid(row=8, columnspan=2)

# Label to display the prediction result
label_result = Label(master, text="Car Purchase Amount: ", fg="blue")
label_result.grid(row=9, columnspan=2)

# Start the Tkinter main loop
master.mainloop()
```
